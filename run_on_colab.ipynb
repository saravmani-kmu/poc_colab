# POC: GitHub → Colab → Google Drive

This notebook:
1. Pulls code from GitHub
2. Runs the Python script to generate output files
3. Zips the output
4. Pushes the zip to Google Drive

## Step 1: Mount Google Drive

```python
from google.colab import drive
drive.mount('/content/drive')
```

## Step 2: Clone the GitHub Repository

**Update the repo URL below with your actual GitHub repo.**

```python
import os

GITHUB_REPO_URL = "https://github.com/saravmani-kmu/poc_colab.git"
BRANCH = "main"

REPO_DIR = "/content/poc_colab"

# Clean previous clone if exists
if os.path.exists(REPO_DIR):
    !rm -rf {REPO_DIR}

!git clone --branch {BRANCH} --single-branch {GITHUB_REPO_URL} {REPO_DIR}
os.chdir(REPO_DIR)
print(f"\nWorking directory: {os.getcwd()}")
!ls -la
```

## Step 3: Run the Python Script

```python
!python main.py
```

## Step 4: Copy the Zip to Google Drive

```python
import shutil
from datetime import datetime

# Destination folder in Google Drive
DRIVE_OUTPUT_DIR = "/content/drive/MyDrive/colab_poc_output"
os.makedirs(DRIVE_OUTPUT_DIR, exist_ok=True)

# Copy with timestamp so runs don't overwrite each other
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
src = "output_bundle.zip"
dst = os.path.join(DRIVE_OUTPUT_DIR, f"output_bundle_{timestamp}.zip")

shutil.copy2(src, dst)
print(f"Uploaded to Google Drive: {dst}")
print(f"File size: {os.path.getsize(dst)} bytes")
```

## Step 5: Verify

```python
# List all outputs in Drive folder
print("Files in Google Drive output folder:")
for f in sorted(os.listdir(DRIVE_OUTPUT_DIR)):
    size = os.path.getsize(os.path.join(DRIVE_OUTPUT_DIR, f))
    print(f"  {f} ({size} bytes)")
```
